```python
import uuid
import json
from pathlib import Path

def save_file_with_uuid(dir, content, suffix='.json', binary=False):
    path = Path(dir)
    
    path.mkdir(parents=True, exist_ok=True)
    
    unique_name = f"{uuid.uuid4().hex}{suffix}"
    unique_path = path / unique_name
    
    if binary:
        unique_path.write_bytes(content)
    else:
        unique_path.write_text(content, encoding='utf-8')
    
    return str(unique_path)
```

```python
client = build_gigachat()
structured_client = client.with_structured_output(TestCase, method="format_instructions")
```

```python
output_dir = "../data/generated_questions"
amount = 100
```

```python
for i in range(amount):
    print(f"{i + 1} / {amount}", end="")
    try:
        response = structured_client.invoke(create_prompt(topics))
        assert isinstance(response, TestCase)
        save_file_with_uuid(output_dir, response.model_dump_json())
        print(f" done", end="")
    except Exception as e:
        print(f"Ошибка: {e}", end="")
    print()
```
